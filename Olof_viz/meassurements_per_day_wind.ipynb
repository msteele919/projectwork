```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_gbg = pd.read_pickle('../Dataframes/df_wind_gbg.pkl')
```

```python
years_gbg = df_gbg['Year'].nunique()
years_gbg
```

```python
sns.histplot(df_gbg, x='Year')
plt.title('Mätningar per år, Göteborg')
plt.ylabel('Antal')
plt.xlabel('År')
plt.show()
```
