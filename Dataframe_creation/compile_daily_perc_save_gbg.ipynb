```python
import pandas as pd
```

```python
df = pd.read_csv('../data/smhi_perc_data_save.csv', usecols=[0, 1, 2, 3, 4], header=6, sep=';', parse_dates=True)
```

```python
df
```

```python
df = df.drop(columns=['Från Datum Tid (UTC)', 'Till Datum Tid (UTC)'], axis=1)
```

```python
df['Representativt dygn'] = pd.to_datetime(df['Representativt dygn'])
```

```python
df['Year'] = df['Representativt dygn'].dt.year
df['Month'] = df['Representativt dygn'].dt.month
```

```python
df
```

```python
monthly_averages = df.groupby(['Year', 'Month'])['Nederbördsmängd'].mean().round(2).reset_index()
```

```python
monthly_averages
```

```python
monthly_averages = monthly_averages.rename({'Nederbördsmängd': 'Monthly avg'}, axis=1)
```

```python
df = df.merge(monthly_averages, how='left', on=['Year', 'Month'])
```

```python
df
```

```python
last_day = df['Representativt dygn'][-1:]
```

```python
last_day = str(last_day.values[0])[0:10]
```

```python
df.head(31)['Nederbördsmängd'].mean()
```

```python
df.to_csv('../data/Olof_percipitation.csv', index=False, mode='w')
```

```python
df_gbg = pd.read_csv('../data/smhi_percipitation_data.csv', usecols=[0, 1, 2, 3, 4], header=11, sep=';')
```

```python
df_gbg = df_gbg.drop(columns=['Från Datum Tid (UTC)', 'Till Datum Tid (UTC)'], axis=1)
```

```python
df_gbg['Representativt dygn'] = pd.to_datetime(df_gbg['Representativt dygn'])
df_gbg['Year'] = df_gbg['Representativt dygn'].dt.year
df_gbg['Month'] = df_gbg['Representativt dygn'].dt.month
```
