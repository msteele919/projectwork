```python
ax = wr.WindroseAxes.from_ax()
ax.bar(df['Vindriktning'], df['Snitt_per_dag'], normed=True, opening=0.8)
plt.show()
```

```python
df_pre_1970 = df[df['Year'] < 1970]
df_post_1970 = df[df['Year'] >= 1970]
```

```python
ax = wr.WindroseAxes.from_ax()
ax.bar(df_pre_1970['Vindriktning'], df_pre_1970['Snitt_per_dag'], normed=True, opening=0.8)
plt.title('Wind directions and speeds, Pre 1970')
ax.legend()
plt.show()
ax = wr.WindroseAxes.from_ax()
ax.bar(df_post_1970['Vindriktning'], df_post_1970['Snitt_per_dag'], normed=True, opening=0.8)
plt.title('Wind Directions and speeds, 1970 ->')
ax.legend()
plt.show()
```
