```python
import pandas as pd
```

```python
df_save = pd.read_csv('../data/smhi_snow_depth_save.csv', header=7, usecols=[0, 1, 2, 3, 4, 5], sep=';')
```

```python
df_save['Datum'] = pd.to_datetime(df_save['Datum'])
df_save['Month'] = df_save['Datum'].dt.month
df_save['Year'] = df_save['Datum'].dt.year
```

```python
df_save
```

```python
df_save['Snittdjup_per_dag_för_månaden'] = df_save.groupby(['Year', 'Month'])['Snödjup'].transform('mean').round(2)
```

```python
df_save.rename(columns=({'Tid (UTC)': 'Tid_UTC', 'Markens tillstånd': 'Markens_tillstånd', 'Kvalitet.1': 'Kvalitet_1'}), inplace=True)
```
