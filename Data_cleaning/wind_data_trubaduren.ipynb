```python
import pandas as pd
```

```python
df_wind_trubaduren = pd.read_csv('../data/smhi_wind_data_trubaduren.csv', sep=';', usecols=[0, 1, 2, 3, 4, 5], header=7)
```

```python
df_wind_trubaduren = df_wind_trubaduren.rename(columns=({'Tid (UTC)': 'Tid_UTC', 'Kvalitet.1': 'Kvalitet_1'}))
```

```python
df_wind_trubaduren
```

```python
df_wind_trubaduren['Datum'] = pd.to_datetime(df_wind_trubaduren['Datum'])
```

```python
df_wind_trubaduren['Vindhastighet_snitt'] = df_wind_trubaduren.groupby('Datum')['Vindhastighet'].transform('mean').round(2)
df_wind_trubaduren['Högsta_vindhastighet'] = df_wind_trubaduren.groupby('Datum')['Vindhastighet'].transform('max').round(2)
df_wind_trubaduren['Year'] = df_wind_trubaduren['Datum'].dt.year
df_wind_trubaduren['Month'] = df_wind_trubaduren['Datum'].dt.month
```
