```python
import pandas as pd
```

Import and clean precipitation data, GBG

```python
df_gbg = pd.read_csv('../data/smhi_percipitation_data_gbg.csv', usecols=[0, 1, 2, 3, 4], header=11, sep=';')
```

```python
df_gbg
```

```python
df_gbg.drop(columns=(['Från Datum Tid (UTC)', 'Till Datum Tid (UTC)']), inplace=True)
```

```python
df_gbg['Representativt dygn'] = pd.to_datetime(df_gbg['Representativt dygn'])
```

```python
df_gbg.rename(columns=({'Representativt dygn': 'Datum'}), inplace=True)
df_gbg['Year'] = df_gbg['Datum'].dt.year
df_gbg['Month'] = df_gbg['Datum'].dt.month
```

```python
df_gbg
```

Import and clean precipitation data, Säve

```python
df_save = pd.read_csv('../data/smhi_precipitation_data_save.csv', usecols=[0, 1, 2, 3, 4], header=6, sep=';')
```

```python
df_save
```

```python
df_save.drop(columns=(['Från Datum Tid (UTC)', 'Till Datum Tid (UTC)']), inplace=True)
df_save['Representativt dygn'] = pd.to_datetime(df_save['Representativt dygn'])
```

```python
df_save.rename(columns=({'Representativt dygn': 'Datum'}), inplace=True)
df_save['Year'] = df_save['Datum'].dt.year
df_save['Month'] = df_save['Datum'].dt.month
```

```python
df_save
```

Make monthly precipitation data


```python
df_monthly_gbg = df_gbg
```

```python
df_monthly_gbg['Avg_nederbörd'] = df_monthly_gbg.groupby(['Year', 'Month'])['Nederbördsmängd'].transform('mean').round(2)
df_monthly_gbg['Max_uppmätta_nederbörd'] = df_monthly_gbg.groupby(['Year', 'Month'])['Nederbördsmängd'].transform('max').round(2)
```

```python
df_monthly_gbg = df_monthly_gbg.groupby(['Year', 'Month']).max().reset_index().drop(columns=('Datum'))
```

```python
df_monthly_save = df_save
```

```python
df_monthly_save['Avg_nederbörd'] = df_monthly_save.groupby(['Year', 'Month'])['Nederbördsmängd'].transform('mean').round(2)
df_monthly_save['Max_uppmätta_nederbörd'] = df_monthly_save.groupby(['Year', 'Month'])['Nederbördsmängd'].transform('max').round(2)
```

```python
df_monthly_save = df_monthly_save.groupby(['Year', 'Month']).max().reset_index().drop(columns=('Datum'))
```

```python
# df_monthly_gbg.to_csv('../data/prepared_data/monthly_precipitation_gbg.csv', index=False)
# df_monthly_gbg.to_pickle('../Dataframes/df_monthly_precipitation_gbg.pkl')

# df_monthly_save.to_csv('../data/prepared_data/monthly_precipitation_save.csv', index=False)
# df_monthly_save.to_pickle('../Dataframes/df_monthly_precipitation_save.pkl')
```

Combine the dataframes covering the entire time period

Combined daily precipitation

```python
last_day_save = df_save['Datum'][-1:]
```

```python
last_day_save = str(last_day_save.values[0])[:10]
```

```python
gbg_daily_limited = df_gbg[df_gbg['Datum'] > last_day_save]
```

```python
compiled_daily_precipitation = pd.concat([df_save, gbg_daily_limited])
```

```python
compiled_daily_precipitation = compiled_daily_precipitation.rename(columns={'Avg_nederbörd': 'Avg_nederbörd_per_dag'})
```

```python
compiled_daily_precipitation.drop(columns=('Nederbördsmängd'), inplace=True)
```

```python
# compiled_daily_precipitation.to_csv('../data/prepared_data/compiled_daily_precipitation_gbg_save.csv', mode='w', index=False)
# compiled_daily_precipitation.to_pickle('../Dataframes/df_compiled_daily_precipitation_gbg_save.pkl')
```

Combined monthly pretticipation

```python
last_year_save, last_month_save = df_monthly_save['Year'][-1:], df_monthly_save['Month'][-1:]
```

```python
last_year_save = int(last_year_save.values[0])
last_month_save = int(last_month_save.values[0])
```

```python
gbg_monthly_limited = df_monthly_gbg[((df_monthly_gbg['Year'] >= last_year_save) & (df_monthly_gbg['Month'] > last_month_save))
                                        | (df_monthly_gbg['Year'] > last_year_save)]
```

```python
gbg_monthly_limited
```

```python
compiled_monthly_precipitation = pd.concat([df_monthly_save, gbg_monthly_limited])
```

```python
compiled_monthly_precipitation.drop(columns=('Nederbördsmängd'), inplace=True)
```

```python
compiled_monthly_precipitation.rename(columns={'Avg_nederbörd': 'Avg_nederbörd_per_dag'}, inplace=True)
```
