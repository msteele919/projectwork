```python
import pandas as pd
```

Clean wind data for GBG

```python
df_gbg = pd.read_csv('../data/smhi_wind_data_gbg.csv', sep=';', usecols=[0, 1, 2, 3, 4, 5], header=12)
```

```python
df_gbg.rename(columns=({'Tid (UTC)': 'Tid_UTC', 'Kvalitet.1': 'Kvalitet_1'}), inplace=True)
```

```python
df_gbg['Datum'] = pd.to_datetime(df_gbg['Datum'])
df_gbg['Year'] = df_gbg['Datum'].dt.year
df_gbg['Month'] = df_gbg['Datum'].dt.month
```

```python
df_gbg['Snitt_per_dag'] = df_gbg.groupby('Datum')['Vindhastighet'].transform('mean').round(2)
df_gbg['Max_uppmätta_hastighet'] = df_gbg.groupby('Datum')['Vindhastighet'].transform('max')
```
