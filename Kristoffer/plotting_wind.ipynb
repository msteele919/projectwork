# Varför blåser det mindre?

Är det detsamma med vind-data utomskärs där ingenting står i vägen?

```python
windyvinga
```

```python
fig, ax = plt.subplots(figsize=(20, 6))
plt.title('Högst uppmätta vindhastighet utomskärs')
plt.ylabel('Vindhastighet i meter per sekund')
plt.xlabel('År')
plt.legend(loc = 'upper right', handles = ['Vinga', 'Trubaduren'], labels=['Vinga', 'Trubaduren'])
sns.lineplot(data = windyvinga, x = 'Datum', y = 'Max_uppmätta_hastighet')
sns.lineplot(data=windytrubadur, x='Datum', y='Högsta_vindhastighet')
# plt.savefig('../plottar/vindhastighetutomskars.png')
```

Som vi ser från denna datan tagen utomskärs, bortsett från den uteblivna datan så är vindhastigheten ungefär densamma över tid.
Blå = Vinga, 
Orange = Trubaduren

```python
orkanvinga = windyvinga.query("Max_uppmätta_hastighet >= 32.6")
svarstormvinga = windyvinga.query("Max_uppmätta_hastighet >= 28.5 and Max_uppmätta_hastighet <=32.5")
stormvinga = windyvinga.query("Max_uppmätta_hastighet >=24.5 and Max_uppmätta_hastighet <= 28.4")
```

```python
print(stormvinga['Max_uppmätta_hastighet'].max())
print(svarstormvinga['Max_uppmätta_hastighet'].max())
print(orkanvinga['Max_uppmätta_hastighet'].max())
```

Då ska vi bara ta reda på när dessa dagarna är.

```python
stormvinga['Datum'].unique()
```

```python
stormvinga = stormvinga.drop_duplicates(subset='Datum')

# Create bins for 20-year intervals
bins = list(range(1920, 2040, 20))  # Adjust the start and end years as needed

# Group the data by 20-year intervals and count observations in each interval
interval_counts = stormvinga.groupby(pd.cut(stormvinga['Year'], bins=bins), observed=True).size()

# Print the result
print(interval_counts)
```

```python
svarstormvinga = svarstormvinga.drop_duplicates(subset='Datum')

# Create bins for 20-year intervals
bins1 = list(range(1920, 2040, 20))  # Adjust the start and end years as needed

# Group the data by 20-year intervals and count observations in each interval
interval_counts1 = svarstormvinga.groupby(pd.cut(svarstormvinga['Year'], bins=bins1), observed=True).size()

# Print the result
print(interval_counts1)
```
